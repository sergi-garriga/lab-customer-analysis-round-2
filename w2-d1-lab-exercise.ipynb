```python
import pandas as pd
```

```python
df = pd.read_csv("files_for_lab/csv_files/marketing_customer_analysis.csv", index_col=[0])
df
```

```python
# Show the dataframe shape.
print(df.shape)
```

```python
# Standardize header names.
df.columns = df.columns.str.upper()
df
```

```python
# Which columns are numerical?
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

numerics_df = df.select_dtypes(include=numerics)
numerics_df.columns
```

```python
# Which columns are categorical?
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

numerics_df = df.select_dtypes(exclude=numerics)
numerics_df.columns
```

```python
# Check and deal with NaN values.
missing_df = pd.DataFrame(df.isna().sum(), columns=["sum of Nan"])
missing_df["% missing"] = (missing_df["sum of Nan"] * 100 / 10909).round(2)
missing_df
```

```python
df['VEHICLE TYPE'].value_counts()
```

```python
# Will drop the Nan values except for the Vehicle Type, where will assume another category "other".
```

```python
df.dropna(subset=['STATE', 'RESPONSE', 'MONTHS SINCE LAST CLAIM', 'NUMBER OF OPEN COMPLAINTS', 'VEHICLE CLASS', 'VEHICLE SIZE'], inplace=True)
df.reset_index(drop=True)
df
```

```python
df['VEHICLE TYPE'].fillna('Other', inplace=True)
df
```
